```python
#change names of mass.download mseed files by julian day 

from obspy import read, read_inventory
import os
counter = 0
path = r"/gps/MSnoise_S/waveforms"
files = []
x = range(1,31)
for i in x:
    for file_name in os.listdir(path):
        n = "__202201%02dT000000Z__202201%02dT000000Z" % (i, i+1)
#        print(n)
        if n in os.path.join(path,file_name):
            old_name = os.path.join(path,file_name)
            new_name = (os.path.join(path,file_name)).replace(n, ".2022.%02d.raw") % (i)
            counter += 1
            files.append (new_name)
            os.rename (old_name, new_name)
for file_name in os.listdir(path):
    if "__20220131T000000Z__20220201T000000Z" in os.path.join(path,file_name):
            old_name = os.path.join(path,file_name)
            new_name = (os.path.join(path,file_name)).replace("__20220131T000000Z__20220201T000000Z", ".2022.31.raw")
            counter += 1
            files.append (new_name)
            os.rename (old_name, new_name)
print (counter)
print (files)
```
